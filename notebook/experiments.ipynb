```python
from langchain.document_loaders import PyPDFLoader
```

```python
import os 
```

```python
file_path = os.path.join(os.getcwd(),"data_sample","sample.pdf")
```

```python
loader = PyPDFLoader(file_path)
```

```python
documents = loader.load()
```

```python
split_doc = RecursiveCharacterTextSplitter(chunk_size = 400, chunk_overlap = 150)
```

```python
docs  = split_doc.split_documents(documents)
```

```python
docs
```

```python
len(docs)
```

```python
from langchain.vectorstores import FAISS
```

```python
vectorstore = FAISS.from_documents(docs,embedding=embed1)
```

```python
relevant = vectorstore.similarity_search("what is Transformer")  
```

```python
prompt_template = """ Answer the question which comes as a input as properly.
if you dont know have suffiecent infromation in the context say sorry i dont know about this question
context : {context}
Question : {question}
Answer:"""
```

```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(template = prompt_template,
               input_variables=["context","question"])
```

```python
prompt
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough
parser = StrOutputParser()

def format_docs(docs):
      return "\n\n".join([doc.page_content for doc in docs ])

retriver = vectorstore.as_retriever()
```

```python
chain =(
    {"context": retriver | format_docs, "question": RunnablePassthrough()}
    | prompt
    | model1
    | parser)
```

```python
output = chain.invoke("what is llm")
```
